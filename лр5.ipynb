```python
# 2.9) создание нового поля вычисленного на основе значений других полей:
# 2.9.1) через выражение на базе имеющихся колонок
data['Category'] = pd.cut(data['Age'], bins=[0, 12, 18, 100], labels=['Children', 'Teenager', 'Adult'])
data
```

```python
# 2.9.2) через DataFrame.apply
def get_category(row):
    if row['Age'] < 12:
        return 'Children'
    elif row['Age'] >= 12 and row['Age'] <= 18:
        return 'Teenage'
    else:
        return 'Adult'
    
data['Category'] = data.apply(get_category, axis=1)
data
```

```python
# 2.9.3) через Series.apply
data['Age'] = data['Age'].astype(int)
cat = list(data['Age'])
a = pd.Series(cat)

def get_cat(age):
    if age < 12:
        return 'Children'
    elif age >= 12 and age < 18:
        return 'Teenage'
    else:
        return 'Adult'

a.apply(get_cat)
```

```python
# 2.10) сортировка по какому-либо из полей
# 2.11) вычислить несколько статистик по колонкам.

mean_age = data['Age'].mean()
min_age = data['Age'].min()
max_age = data['Age'].max()
print("Среднее значение: ", mean_age)
print("Минимальное значение: ", min_age)
print("Максимальное значение: ", max_age)
```
